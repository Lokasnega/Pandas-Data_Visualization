###### #Average completion of python course or my_sql or python english or computational thinking

```python
df['mean'] = df.iloc[:,[1,2,7,8]].mean(axis =1)
df.drop(["Previous Geekions","CodeKata Score",'Department','Rising'],axis = "columns")
```

```python
avg = df.loc[:,["python","mysql","python_en","computational_thinking"]]
avg.mean()
```

###### rising star of the week (top 3 candidate who performed well in that particular week)

```python
df.drop(["Previous Geekions",'Department','Rising',
         'python','mysql','python_en','computational_thinking','mean'],axis = "columns").sort_values(by="CodeKata Score",ascending = False).head(3)
```

###### Shining stars of the week (top 3 candidates who has highest geekions)

```python
df['Total_Geekions'] = df['Previous Geekions'] + df['CodeKata Score']
df.drop(["Previous Geekions","CodeKata Score",'Department','Rising',
         'python','mysql','python_en','computational_thinking','mean'],axis = "columns").sort_values(by="Total_Geekions",ascending = False).head(3)
```

###### Department wise codekata performence (pie chart)

```python
import matplotlib.pyplot as plt
df.groupby(['Department']).sum().plot(kind='pie', y='Total_Geekions',autopct='%1.0f%%', shadow = True, startangle = 140)
```
